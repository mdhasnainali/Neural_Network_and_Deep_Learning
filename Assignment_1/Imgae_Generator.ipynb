```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./evaluation_metrics.csv")

df = df.head(10)
df
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Evaluation Accuracy', y='Model Name', data=df.sort_values("Evaluation Accuracy", ascending=False))
plt.title("Model Accuracy on 20 Classes of CIFAR-100")
plt.xlabel("Evaluation Accuracy")
plt.ylabel("Model Name")
plt.tight_layout()
plt.savefig("analysis/accuracy_vs_model_name.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Evaluation Time (Sec)', y='Evaluation Accuracy', hue='Model Name')
plt.title("Accuracy vs Evaluation Time")
plt.xlabel("Evaluation Time (Seconds)")
plt.ylabel("Evaluation Accuracy")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("analysis/accuracy_vs_eval_time.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Model Size (MB)', y='Evaluation Accuracy', hue='Model Name')
plt.title("Accuracy vs Model Size")
plt.xlabel("Model Size (MB)")
plt.ylabel("Evaluation Accuracy")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("analysis/accuracy_vs_model_size.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Training Time (Sec)', y='Evaluation Accuracy', hue='Model Name')
plt.title("Accuracy vs Training Time")
plt.xlabel("Training Time (Seconds)")
plt.ylabel("Evaluation Accuracy")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("analysis/accuracy_vs_training_time.png")
plt.show()
```
